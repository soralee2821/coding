# More Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Global API

```python
x = np.arange(-10,11)
```

```python
plt.figure(figsize=(12,6))

plt.title('My nice plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12,6))

plt.title('My nice plot')

plt.subplot(1, 2, 1)
plt.plot(x, x ** 2)
plt.plot([0, 0, 0],[-10, 0, 100])
plt.legend(['x^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10],[-50, -50, -50])
plt.legend(['-x^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X squared')
```
