# More Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Global API

```python
x = np.arange(-10,11)
```

```python
plt.figure(figsize=(12,6))

plt.title('My nice plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12,6))

plt.title('My nice plot')

plt.subplot(1, 2, 1)
plt.plot(x, x ** 2)
plt.plot([0, 0, 0],[-10, 0, 100])
plt.legend(['x^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10],[-50, -50, -50])
plt.legend(['-x^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X squared')
```

### Object Oriented Programming (OOP) Interface

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x**2), color='red', linewidth=3, marker='o', markersize=8, label='X^2'
)
axes.plot(x, -1 * (x**2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title('My nice plot')

axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted')

axes.set_title('My nice plot')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label='solid green')
axes.plot(x, x + 1, '--c', label='dashed cyan')
axes.plot(x, x + 2, '-.b', label='dashdot blue')
axes.plot(x, x + 3, ':r', label='dotted red')

axes.set_title('My nice plot')
```

### Scatter plot

```python
# rand : create random value from standard normal distribution in a given shape
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

### Histogram

```python
values = np.random.randn(1000)
```

```python
plt.subplots(figsize=(12,6))

plt.hist(values, bins=100, alpha=0.8, histtype='bar', 
         color='steelblue', edgecolor='green')

plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
fig.savefig('hist.png')
```

### KDE (Kernel Density Estimation)

```python
from scipy import stats
```

```python
density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12,6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Combine plots

```python
plt.subplots(figsize=(12,6))

plt.hist(values, bins=100, alpha=0.8, density=1,
         histtype='bar', color='steelblue', edgecolor='green')

plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

### Bar plots

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize=(12,4))

barWidth = 0.5

plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

plt.show()
```

```python
plt.figure(figsize=(12,4))

barWidth = 0.5

plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e15055', bottom=Y, label='Label Y2')

plt.show()
```
